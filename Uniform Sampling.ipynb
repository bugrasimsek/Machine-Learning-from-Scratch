```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
#use degree to radian converter first
x = math.radians(90)

#then get the sin of x
math.sin(x)
```

```python
#get a bunch of random degrees between 0 - 360
sampleSize = 20 
sinList = list() #sin(x) values [0,1]
degreeList = list() #xÂ° values [0,360]

lowerBound = 0
upperBound = 360

for i in range (sampleSize):
    randDegree = random.randint(lowerBound,upperBound) #get x, some random number between 0 - 360
    degreeList.append(randDegree)
    
    randSin = math.sin(math.radians(randDegree)) #get sample value, which is sin(radian(x))
    sinList.append(randSin)
```

```python
print("Random x values: " + str(degreeList))
print("Corresponding sin(x) values: " + str(sinList))
```

```python
plt.scatter(degreeList, sinList)
plt.show()
```

```python
#add some noise to sample list to use them as target values
sampleList = list()



for sin in sinList:
    #noise is random between [0.10, 0.20]
    noise = random.randint(10,20) / 100
    
    #and noise is sometimes randomly negative
    noise = noise * random.choice([-1,1])
    
    sample = sin + noise
    sampleList.append(sample)
```

```python
#convert degrees to radians to plot with sin(x)
radList = [math.radians(deg) for deg in degreeList]
```

```python
#Plot sin(x) using numpy
x = np.arange(0, 2 * np.pi, 0.1) #start (0), end (2Ï), step (0.1 radians)
y = np.sin(x)
 
# Show sin values (without noise)
plt.plot(x, y, color='red')
plt.scatter(radList, sinList)
plt.show()
```

```python
# Show sample values (sin with noise)
plt.plot(x, y, color='red')
plt.scatter(radList, sampleList, color="green")
plt.show()
```
